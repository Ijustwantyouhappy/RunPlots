注意：`X_pca` is the matrix of the transformed components from X，shape = `[n_samples, n_components]`，如果参数`X_pca`不传入数据的话，the function computes PCA automatically using `mlxtend.feature_extraction.PrincipalComponentAnalysis`，Expected `n_componentes >= max(dimensions)`

```python
from sklearn.datasets import load_iris
import numpy as np
from mlxtend.plotting import plot_pca_correlation_graph
```

```python
iris = load_iris()
X = iris.data
y = iris.target
cols = iris.feature_names
```

```python
figure, correlation_matrix = plot_pca_correlation_graph(
    X, cols, dimensions=(1, 2), figure_axis_size=10
)
```

```python
figure, correlation_matrix = plot_pca_correlation_graph(
    X, cols, dimensions=(1, 4), figure_axis_size=10
)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X)
```

<font color="red">传入`X_pca`必须同时传入`explained_variance`，而且`dimensions`第二个位置只能设置最大维，很不好用。</font>

```python
figure, correlation_matrix = plot_pca_correlation_graph(
    X, cols, dimensions=(1, 4), X_pca=X_pca, 
    explained_variance=pca.explained_variance_,
    figure_axis_size=10
)
```
