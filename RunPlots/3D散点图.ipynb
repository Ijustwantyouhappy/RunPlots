```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pyecharts.options as opts
from pyecharts.charts import Scatter3D

sns.set(style='white')


def scatter_3d(df, x, y, z, module='matplotlib'):
    if module in ['matplotlib', 'seaborn']:
        fig = plt.figure()
        ax = Axes3D(fig, rect=[0, 0.1, 1.5, 2], azim=134, elev=48)
        ax.scatter(df[x], df[y], df[z])
        ax.set_xlabel(x)
        ax.set_ylabel(y)
        ax.set_zlabel(z);
    elif module == 'pyecharts':
        obj = Scatter3D(
            init_opts=opts.InitOpts()
        ).add(
            series_name="",
            data=df[[x, y, z]].values.tolist(),
            xaxis3d_opts=opts.Axis3DOpts(
                name=x,
                type_="value",
                # textstyle_opts=opts.TextStyleOpts(color="#fff"),
            ),
            yaxis3d_opts=opts.Axis3DOpts(
                name=y,
                type_="value",
                # textstyle_opts=opts.TextStyleOpts(color="#fff"),
            ),
            zaxis3d_opts=opts.Axis3DOpts(
                name=z,
                type_="value",
                # textstyle_opts=opts.TextStyleOpts(color="#fff"),
            ),
            grid3d_opts=opts.Grid3DOpts(width=100, height=100, depth=100)
        ).set_global_opts(
                visualmap_opts=[]
        ).render_notebook()
        return obj
```

```python
import numpy as np
import pandas as pd
 
# x, y, z 均为 0 到 1 之间的 100 个随机数
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)
df = pd.DataFrame({'x': x, 'y': y, 'z': z})
```

```python
scatter_3d(df, 'x', 'y', 'z')
```

```python
scatter_3d(df, 'x', 'y', 'z', module='pyecharts')
```
