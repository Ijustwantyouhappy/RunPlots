# `axhline` 和 `axvline`
相对位置，0是图的最左侧，1是图的最右侧。

```python
?plt.axhline
```

```python
plt.plot(range(10), range(10))
plt.axhline(y=5, xmin=0.1, xmax=0.9)
plt.show();
```

# `hlines` 和 `vlines`
绝对位置，`xmin`和`xmax`传入的是具体坐标。

```python
plt.plot(range(10), range(10))
plt.hlines(y=5, xmin=4, xmax=6)  # 自动换了颜色
plt.hlines(y=2, xmin=2, xmax=8, linestyle='--', color='y')
plt.show();
```

# `plt.plot`
能替代`hlines`和`vlines`，但似乎不能替代`axhline`和`axvline`。

```python
plt.plot(range(10), range(10))
plt.plot([2, 8], [5, 5])
plt.show();
```
