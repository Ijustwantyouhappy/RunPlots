# 显示图片

```python
%matplotlib inline

import scipy.misc
import matplotlib.pyplot as plt

face = scipy.misc.face()

plt.figure()
plt.axis('off')
plt.imshow(face)
plt.show();
```

# 显示灰度图
从下图中可以看出， `r` 表示 `reverse`，代表黑白反过来

```python
plt.figure(figsize=(10, 6))

# 注意，灰度图的传入图片只能有一个通道
if face.ndim == 3:
    face = face[:, :, 0]
plt.subplot(221); plt.axis('off'); plt.imshow(face, cmap=plt.cm.gray)
plt.subplot(222); plt.axis('off'); plt.imshow(face, cmap=plt.cm.gray_r)
plt.subplot(223); plt.axis('off'); plt.imshow(face, cmap='gray')
plt.subplot(224); plt.axis('off'); plt.imshow(face, cmap='gray_r')
plt.show();
```
